```python
import pandas as pd
import numpy as np

print("1. Seleccionando columnas con los nombres correctos...")
# Pasamos todas las columnas a minúsculas
df_enoe.columns = df_enoe.columns.str.lower()

# CAMBIO CLAVE: Usamos 'cve_mun' y 'cvegeo' en lugar de 'ent' y 'mun'
columnas_disponibles = ['cvegeo', 'cve_mun', 'eda', 'sex', 'clase2', 'fac_tri'] 
df_limpio = df_enoe[columnas_disponibles].copy()

print("2. Separando el Estado (ent) desde la clave geográfica...")
# Como 'cvegeo' junta estado y municipio (ej. 30123), si lo dividimos entre 1000 
# nos da el número del estado exacto (30).
df_limpio['ent'] = df_limpio['cvegeo'] // 1000

# Renombramos 'cve_mun' a 'mun' para no cambiar el resto del código
df_limpio = df_limpio.rename(columns={'cve_mun': 'mun'})

print("3. Filtrando solo los estados del derrame (Tabasco=27, Veracruz=30)...")
df_limpio = df_limpio[df_limpio['ent'].isin([27, 30])]

print("4. Traduciendo los códigos del INEGI a texto real...")
dic_sexo = {1: 'Hombre', 2: 'Mujer'}
dic_ocupacion = {1: 'Trabaja', 2: 'No Trabaja (Desocupado)', 3: 'No Activo', 4: 'No Especificado'}

df_limpio['Sexo'] = df_limpio['sex'].map(dic_sexo)
df_limpio['Estatus_Laboral'] = df_limpio['clase2'].map(dic_ocupacion)

print("5. Filtrando para excluir a menores y jubilados/estudiantes...")
df_pea = df_limpio[df_limpio['clase2'].isin([1, 2])].copy()

# Convertimos la edad a número
df_pea['eda'] = pd.to_numeric(df_pea['eda'], errors='coerce')

print("\n--- BASE DE DATOS LIMPIA (MUESTRA) ---")
display(df_pea[['ent', 'mun', 'eda', 'Sexo', 'Estatus_Laboral', 'fac_tri']].head())

# =====================================================================
# OBJETIVO: ¿Cuántos trabajan y cuántos no por Municipio y Género?
# =====================================================================
print("\n--- RESUMEN: TRABAJADORES VS DESEMPLEADOS ---")

# Agrupamos y SUMAMOS el factor de expansión (fac_tri)
resumen_empleo = df_pea.groupby(['ent', 'mun', 'Sexo', 'Estatus_Laboral'])['fac_tri'].sum().reset_index(name='Total_Personas')

# Redondeamos
resumen_empleo['Total_Personas'] = resumen_empleo['Total_Personas'].round(0).astype(int)

display(resumen_empleo.head(10))

# =====================================================================
# OBJETIVO EXTRA: Edades de las personas que NO trabajan
# =====================================================================
print("\n--- EDADES PROMEDIO DE DESEMPLEADOS POR MUNICIPIO ---")

df_desempleados = df_pea[df_pea['Estatus_Laboral'] == 'No Trabaja (Desocupado)']

edades_desempleo = df_desempleados.groupby(['ent', 'mun', 'Sexo'])['eda'].mean().reset_index(name='Edad_Promedio')
edades_desempleo['Edad_Promedio'] = edades_desempleo['Edad_Promedio'].round(1)

display(edades_desempleo.head())

# Guardamos
resumen_empleo.to_csv('datos_ods8_veracruz_tabasco.csv', index=False)
print("\n¡Listo! Archivo 'datos_ods8_veracruz_tabasco.csv' guardado exitosamente.")
```

```python
import pandas as pd

print("1. Recuperando los datos de Veracruz y Tabasco...")
# Aseguramos la columna de estado por si reiniciaste el Notebook
df_enoe['ent'] = pd.to_numeric(df_enoe['cvegeo'], errors='coerce') // 1000
df_estados = df_enoe[df_enoe['ent'].isin([27, 30])].copy()

print("\n🔍 2. INVESTIGANDO LA COLUMNA DE EMPLEO...")
print("Valores reales ocultos en 'clase2':", df_estados['clase2'].unique()[:10])

# --- FILTRO A PRUEBA DE BALAS ---
# Convertimos a texto y quitamos cualquier espacio en blanco invisible (.str.strip())
df_estados['clase2_limpia'] = df_estados['clase2'].astype(str).str.strip()

# Buscamos variaciones de 1 y 2 (como texto y como decimal)
df_pea = df_estados[df_estados['clase2_limpia'].isin(['1', '2', '1.0', '2.0'])].copy()

if len(df_pea) > 0:
    print(f"✅ ¡Filtro reparado! Sobrevivieron {len(df_pea)} personas.")
    columna_usar = 'clase2_limpia'
else:
    print("⚠️ 'clase2' está vacía o rota. Activando Plan B (columna emp_ppal)...")
    # emp_ppal: 1 = Trabaja, 2 = No Trabaja
    df_estados['emp_ppal_limpia'] = df_estados['emp_ppal'].astype(str).str.strip()
    df_pea = df_estados[df_estados['emp_ppal_limpia'].isin(['1', '2', '1.0', '2.0'])].copy()
    columna_usar = 'emp_ppal_limpia'
    print(f"✅ Plan B exitoso: Sobrevivieron {len(df_pea)} personas.")

# --- 3. ARMANDO EL RESUMEN FINAL ---
print("\n3. Armando y traduciendo la tabla final...")
# Renombramos el municipio
df_pea = df_pea.rename(columns={'cve_mun': 'mun'})

# Diccionarios de traducción (usando texto para evitar errores de Python)
dic_sexo = {'1': 'Hombre', '2': 'Mujer', '1.0': 'Hombre', '2.0': 'Mujer'}
dic_ocupacion = {'1': 'Trabaja', '2': 'No Trabaja (Desocupado)', '1.0': 'Trabaja', '2.0': 'No Trabaja (Desocupado)'}

df_pea['Sexo'] = df_pea['sex'].astype(str).str.strip().map(dic_sexo)
df_pea['Estatus_Laboral'] = df_pea[columna_usar].map(dic_ocupacion)

# Aseguramos que los números para sumar sean correctos
df_pea['fac_tri'] = pd.to_numeric(df_pea['fac_tri'], errors='coerce').fillna(0)
df_pea['eda'] = pd.to_numeric(df_pea['eda'], errors='coerce')

# Agrupamos los datos
resumen_empleo = df_pea.groupby(['ent', 'mun', 'Sexo', 'Estatus_Laboral'])['fac_tri'].sum().reset_index(name='Total_Personas')
resumen_empleo['Total_Personas'] = resumen_empleo['Total_Personas'].round(0).astype(int)

print("\n--- ¡MISION CUMPLIDA! AQUÍ ESTÁ TU TABLA ---")
display(resumen_empleo.head(10))

# Guardar
resumen_empleo.to_csv('datos_ods8_veracruz_tabasco.csv', index=False)
print("\n¡Archivo final 'datos_ods8_veracruz_tabasco.csv' guardado y listo para graficar!")
```
