Estructura de los diccionarios

Un diccionario es una colección desordenada de elementos almacenados como pares clave-valor, donde las claves deben ser únicas.
Se pueden acceder a los valores utilizando las claves correspondientes, lo que facilita la organización y manipulación de datos.
Ejemplos de uso de diccionarios

Un diccionario puede contener diferentes tipos de datos, como cadenas, listas, números y otros diccionarios.
Se pueden modificar, actualizar y eliminar elementos dentro de un diccionario utilizando métodos como pop.
Iteración y manipulación

Los diccionarios son iterables, lo que permite recorrer sus claves y valores.
Se pueden obtener las claves y valores de un diccionario utilizando los métodos keys() y values().
Importancia en ciencia de datos

Los diccionarios son fundamentales en el proceso de ciencia de datos, ya que pueden ser utilizados para crear DataFrames en pandas, facilitando la manipulación de información.

pop() si no se usar shib + TAB dentro del parentesis

#  Listas en python declaracion y operaciones esenciales
Definición de listas

Una lista es una colección ordenada y mutable de elementos, lo que permite cambiar, agregar y eliminar elementos después de su creación.
Se definen utilizando corchetes y pueden contener diferentes tipos de datos, incluyendo números, cadenas y otras listas.
Creación y acceso a listas

Se puede crear una lista a partir de elementos separados por comas dentro de corchetes, como en el ejemplo de una lista de números.
Los elementos de una lista se pueden acceder mediante su índice, comenzando desde 0, lo que permite obtener el valor correspondiente a una posición específica.
Mutabilidad y orden

Las listas son mutables, lo que significa que sus elementos pueden ser modificados.
Mantienen un orden específico, lo que permite acceder a los elementos de manera predecible.      []
listas apartir de una tupla, mutables y ordenadas cambias elementos,acceso a sus elementos por el indice

```python
# sorted una funcion en medio parentesis nuestra lista  
numeros=[16,4,9,1,3,20,8]
ordenados=sorted(numeros)
print(numeros)
print(ordenados)
```

```python
# sorted  , indicar tipo de ordenamiento descendente 
numeros=[16,4,9,1,3,20,8]
ordenados=sorted(numeros,reverse=True)
print(numeros)
print(ordenados)
```
