En esta sección analizamos el impacto de la industria extractiva en el ecosistema.


```python
# 1. Importar las herramientas necesarias
import pandas as pd
import matplotlib.pyplot as plt

# 2. Crear los datos históricos (Años, Derrames en barriles y Pesca en toneladas)
datos_historicos = {
    'Año': [2018, 2019, 2020, 2021, 2022, 2023, 2024],
    'Derrames_Barriles': [2500, 3100, 2800, 4500, 1200, 1500, 2000],
    'Produccion_Pesca_Ton': [847040, 820000, 750000, 780000, 700000, 680000, 650000]
}

# 3. Convertir los datos en una tabla de Pandas (Dataframe)
df_historia = pd.DataFrame(datos_historicos)

# 4. Mostrar la tabla para confirmar que se cargó bien
print(df_historia)
```

```python
# 1. Crear el lienzo de la gráfica
fig, ax1 = plt.subplots(figsize=(10, 6))

# 2. Configurar el Eje Izquierdo (Barras Negras = Derrames)
color_derrames = '#333333' # Gris oscuro/negro
ax1.bar(df_historia['Año'], df_historia['Derrames_Barriles'], color=color_derrames, alpha=0.5, label='Derrames (Barriles)')
ax1.set_xlabel('Año', fontsize=12, fontweight='bold')
ax1.set_ylabel('Impacto de Derrames (Barriles)', color=color_derrames, fontsize=12)
ax1.tick_params(axis='y', labelcolor=color_derrames)

# 3. Configurar el Eje Derecho (Línea Azul = Pesca)
ax2 = ax1.twinx() # Esto crea un segundo eje Y que comparte el mismo eje X (los años)
color_pesca = '#0055A4' # Azul océano
ax2.plot(df_historia['Año'], df_historia['Produccion_Pesca_Ton'], color=color_pesca, marker='o', linewidth=3, markersize=8, label='Producción Pesquera')
ax2.set_ylabel('Producción Pesquera (Toneladas)', color=color_pesca, fontsize=12)
ax2.tick_params(axis='y', labelcolor=color_pesca)

# 4. Título y diseño visual
plt.title('El Costo del Crudo: Derrames vs. Declive Pesquero en el Golfo', fontsize=14, fontweight='bold', pad=20)
fig.tight_layout() # Ajusta los márgenes automáticamente
plt.grid(axis='y', linestyle='--', alpha=0.3) # Líneas guía suaves

# 5. Mostrar la gráfica terminada
plt.show()
```
